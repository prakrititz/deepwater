```python
from ultralytics import YOLO
model = YOLO("yolov8m.pt")
```

```python
model.train(data='data.yaml', epochs=5)
```

```python
from google.colab import drive        #this saves the weights of the model to the drive
import shutil
import os

# Mount Google Drive
#drive.mount('/content/drive')

# Define the source and destination paths
source_folder = '/content/runs'
destination_folder = '/content/drive/My Drive/Xylem/runs_21august'

# Check if the destination folder already exists
if os.path.exists(destination_folder):
    # Remove the existing folder in Google Drive to avoid conflicts
    shutil.rmtree(destination_folder)

# Copy the updated 'runs' folder to Google Drive
shutil.copytree(source_folder, destination_folder)

# Print a success message
print(f"The 'runs' folder has been successfully updated in {destination_folder} in Google Drive.")
```
